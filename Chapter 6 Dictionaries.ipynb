**6-5.** Rivers: Make a dictionary containing three major rivers and the country each river runs through. One key-value pair might be 'nile': 'egypt'.
- Use a loop to print a sentence about each river, such as The Nile runs through Egypt.
- Use a loop to print the name of each river included in the dictionary.
- Use a loop to print the name of each country included in the dictionary.


```python
rivers = {
    "parana": "argentina",
    "nilo": "egipto",
    "mississippi": "eeuu"
}

for river, country in rivers.items():
    print(f"The {river.title()} runs through {country.title()}")

print("\nLista de Rios:")
for river in rivers:
    print(f"\t{river.title()}")

print("\nLista de Paises:")
for country in rivers.values():
    print(f"\t{country.title()}")
```

**6-11.** Cities: Make a dictionary called cities. Use the names of three cities as keys in your dictionary. The keys for each city’s dictionary should be something like country, population, and fact. Print the name of each city and all of the information you have stored about it.


```python
cities = {
    "rosario": {
        'population': 992594,
        'pais': 'Argentina',
        'curiosidad': 'Aqui se encuentra el monumento a la bandera'
    },
    "edimburgo": {
        'population': 482005,
        'pais': 'Escocia',
        'curiosidad': 'Edimburgo es la ciudad de las chimeneas… pero huele a pan'
    },
    "buenos aires": {
        'population': 2890151,
        'pais': 'Argentina',
        'curiosidad': 'El subterráneo porteño fue el primero de Sudamérica'
    }
}

for city, city_info in cities.items():
    print(f"\nCiudad: {city.title()}")
    for fact, info in city_info.items():
        print(f"\t{fact.title()}: {info}")
    print(city, city_info)
```
