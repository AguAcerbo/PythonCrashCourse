**7-1.** Rental Car: Write a program that asks the user what kind of rental car they would like. Print a message about that car, such as “Let me see if I can find you a Subaru.”

```python
def c7_ej1(nombre_ejercicios, ejercicio):
    print("-------------------------------")
    print(f"{nombre_ejercicios[ejercicio]}")
    print("-------------------------------")

    car = input("What kind of rental car would you like?\n").lower()
    print(f"Let me see if I can find you a {car.title()}")
    return None
```

**7-2.** Restaurant Seating: Write a program that asks the user how many people are in their dinner group. If the answer is more than eight, print a message saying they’ll have to wait for a table. Otherwise, report that their table is ready.

```python
def c7_ej2(nombre_ejercicios, ejercicio):
    print("-------------------------------")
    print(f"{nombre_ejercicios[ejercicio]}")
    print("-------------------------------")

    dinner_group = int(input("How many people are in your dinner group?\n"))

    if dinner_group > 8:
        print(f"I'm sorry but you have to wait for a table.")
    else:
        print(f"Please come with me. Your table is ready.")
    return None
```

**7-3.** Multiples of Ten: Ask the user for a number, and then report whether the number is a multiple of 10 or not.

```python
def c7_ej3(nombre_ejercicios, ejercicio):
    print("-------------------------------")
    print(f"{nombre_ejercicios[ejercicio]}")
    print("-------------------------------")

    while True:
        numero = input("Ingrese un numero para evaluarlo o quit para salir del programa.\n")
        if numero.lower() == 'quit':
            break
        else:
            nro = int(numero)
            if int(numero) % 10 == 0:
                print(f"El numero {int(numero)} es multiplo de 10.")
            else:
                print(f"El numero {int(numero)} no es multiplo de 10")
    return None
```

**7-4.** Pizza Toppings: Write a loop that prompts the user to enter a series of pizza toppings until they enter a 'quit' value. As they enter each topping, print a message saying you’ll add that topping to their pizza.

```python
def c7_ej4(nombre_ejercicios, ejercicio):
    print("-------------------------------")
    print(f"{nombre_ejercicios[ejercicio]}")
    print("-------------------------------")

    while True:
        topping = input("Ingrese el topping que desea gregar a su pizza."
                        "\nO 'quit' para salir del programa.\n\t\t")
        if topping.lower() == 'quit':
            break
        else:
            print(f"\t\tSe agregara {topping.upper()} a su pizza.")
    return None
```

**7-5.** Movie Tickets: A movie theater charges different ticket prices depending on a person’s age. If a person is under the age of 3, the ticket is free; if they are between 3 and 12, the ticket is $10; and if they are over age 12, the ticket is $15. Write a loop in which you ask users their age, and then tell them the cost of their movie ticket.

```python
def c7_ej5(nombre_ejercicios, ejercicio):
    print("-------------------------------")
    print(f"{nombre_ejercicios[ejercicio]}")
    print("-------------------------------")

    while True:
        age = input("Ingrese la edad de la persona.\n"
                    "Ingrese QUIT si quiere salir.\n\t\t")
        if age.lower() == 'quit':
            break
        else:
            if int(age) < 3:
                print("The ticket is free.")
            elif int(age) < 12:
                print("The ticket is $10.")
            elif int(age) >= 12:
                print("The ticket is $15.")
    return None
```

```python
def c7_ej7(nombre_ejercicios, ejercicio):
    print("-------------------------------")
    print(f"{nombre_ejercicios[ejercicio]}")
    print("-------------------------------")

    bucles = 1
    while True:
        print(f"{bucles} Bucle infitnito.\n"
              f"Presiones CTRL+F2 para terminar programa.")
        bucles += 1
    return None
```

**7-8.** Deli: Make a list called sandwich_orders and fill it with the names of various sandwiches. Then make an empty list called finished_sandwiches. Loop through the list of sandwich orders and print a message for each order, such as I made your tuna sandwich. As each sandwich is made, move it to the list of finished sandwiches. After all the sandwiches have been made, print a 
message listing each sandwich that was made.

```python
def c7_ej8(nombre_ejercicios, ejercicio):
    print("-------------------------------")
    print(f"{nombre_ejercicios[ejercicio]}")
    print("-------------------------------")

    sandwich_orders = ["pastrami", "bolognesa", "chimichurri","pastrami", "matambre","pastrami", "aceitunas y mayo", "bolognesa", "matambre"]
    finished_sanwiches = []

    while sandwich_orders:
        sandwich = sandwich_orders.pop()
        print(f"He terminado tu sandwich de {sandwich}")
        finished_sanwiches.append(sandwich)
    print("\nSandwiches terminados:")
    for sandwich in finished_sanwiches:
        print(f"\t{sandwich}")
    return None
```

**7-9.** No Pastrami: Using the list sandwich_orders from Exercise 7-8, make sure the sandwich 'pastrami' appears in the list at least three times. Add code near the beginning of your program to print a message saying the deli has run out of pastrami, and then use a while loop to remove all occurrences of 'pastrami' from sandwich_orders. Make sure no pastrami sandwiches end up 
in finished_sandwiches.

```python
def c7_ej9(nombre_ejercicios, ejercicio):
    print("-------------------------------")
    print(f"{nombre_ejercicios[ejercicio]}")
    print("-------------------------------")

    sandwich_orders = ["pastrami", "bolognesa", "chimichurri", "pastrami", "matambre", "pastrami", "aceitunas y mayo",
                       "bolognesa", "matambre"]
    finished_sanwiches = []

    print("There is no more pastrami, I'm sorry.")
    while "pastrami" in sandwich_orders:
        sandwich_orders.remove("pastrami")
    while sandwich_orders:
        sandwich = sandwich_orders.pop()
        print(f"He terminado tu sandwich de {sandwich}")
        finished_sanwiches.append(sandwich)
    print("\nSandwiches terminados:")
    for sandwich in finished_sanwiches:
        print(f"\t{sandwich}")
    return None
```

**7-10.** Dream Vacation: Write a program that polls users about their dream vacation. Write a prompt similar to If you could visit one place in the world, where would you go? Include a block of code that prints the results of the poll.

```python
def c7_ej10(nombre_ejercicios, ejercicio):
    print("-------------------------------")
    print(f"{nombre_ejercicios[ejercicio]}")
    print("-------------------------------")

    respuestas = {}

    poll_active = True

    while poll_active:
        name = input("Cual es su nombre?")
        response = input("Si pudieras elegir un destino para vacacionar, ¿cual seria?")

        respuestas[name] = response

        repeat = input("¿Desea dejar a otra persona realizar la entrevista? (si / no)")
        if repeat.lower() == "no":
            poll_active = False

    print("\n--- Poll results ---")
    for name, response in respuestas.items():
        print(f"A {name.title()} le gustaria ir a {response.title()}")
    return None
```
