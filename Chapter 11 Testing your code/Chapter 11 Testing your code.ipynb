# Chater 11 Test your code

The module unitest fro the Python standar library provides tools for testing your code. A *unit test* verifies that one specific aspect of a function's behaviour is correct. A *test case* is a collection of unit test that together prove that a function behaves as it's supposed to, within the full range of situations you expect it to handle.

Assert Methods Availabel from the unittest Module:

|**Method** | **Use**|
|-----------|--------|
|assertEqual(a, b) | Verify that a == b|
|assertNotEqual(a, b) | Verify that a != b|
|assertTrue(x) | Verify that x is True|
|assertFalse(x) | Verify that x is False|
|assertIn(item, list) | Verify that item is in the list|
|assertNotIn(item, list) | Verify that item is not in the list|
